```python
from sklearn.preprocessing import MaxAbsScaler
from scipy.sparse import csr_matrix
from sklearn.feature_selection import SelectKBest, f_classif  # Changed from chi2 to f_classif as chi square can't handle negative values


X = csr_matrix([[0, -1, 3], [0, 2, 1], [3, -1, 0]])
y = [0, 1, 0]

# MaxAbsScaler works with sparse matrices and handles negative values
X_scaled = MaxAbsScaler().fit_transform(X)

# Changed chi2 to f_classif which can handle negative values
selector = SelectKBest(score_func=f_classif, k=2)
X_new = selector.fit_transform(X, y)
```
