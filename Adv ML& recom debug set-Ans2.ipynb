```python
from sklearn.metrics.pairwise import cosine_similarity
ratings = pd.DataFrame({
    'User1': [5, None, 3],
    'User2': [None, 4, None],
    'User3': [1, 2, None]
})
sim = cosine_similarity(
    ratings.T.apply(lambda x: x.fillna(x.mean()), axis=1)# since none values are there, lambda fills it with mean of existing ratings
    
)
print(sim)
```

```python
#9.
from sklearn.neighbors import NearestNeighbors
import numpy as np

X = np.array([[1, 0, 1], [0, 1, 0], [1, 1, 1]])
model = NearestNeighbors()
model.fit(X)
distances, indices = model.kneighbors([X[0]], n_neighbors=2)
print(indices)
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

X = np.array([[1, 0, 1], [0, 1, 0], [1, 1, 1]])
model = NearestNeighbors()
model.fit(X)
distances, indices = model.kneighbors([X[0]], n_neighbors=3)# no of neighbors need to match array elements
print(indices)
```
